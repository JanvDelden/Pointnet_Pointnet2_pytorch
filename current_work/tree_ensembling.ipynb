# voxelize forest

```python
#@title voxelize forest fn
def data_prepare(points, label, voxel_size=0.04):
    # regularize point density, in voxel of voxel_size, a maximum of one point is allowed
    if voxel_size:
        uniq_idx = voxelize(points, voxel_size)
        points, label = points[uniq_idx], label[uniq_idx]
    return points, label


def voxelize(points, voxel_size=0.05, mode=0):
    discrete_points = np.floor(points / np.array(voxel_size))
    key = fnv_hash_vec(discrete_points)
    idx_sort = np.argsort(key)
    key_sort = key[idx_sort]
    _, count = np.unique(key_sort, return_counts=True)
    if mode == 0:  # train mode
        idx_select = np.cumsum(np.insert(count, 0, 0)[0:-1]) + np.random.randint(0, count.max(), count.size) % count
        idx_unique = idx_sort[idx_select]
        return idx_unique
    else:  # val mode
        return idx_sort, count


def fnv_hash_vec(arr):
    """
    FNV64-1A see wikipedia
    """
    assert arr.ndim == 2
    # Floor first for   points
    arr = arr.copy()
    arr = arr.astype(np.uint64, copy=False)
    hashed_arr = np.uint64(14695981039346656037) * np.ones(arr.shape[0], dtype=np.uint64)
    for j in range(arr.shape[1]):
        hashed_arr *= np.uint64(1099511628211)
        hashed_arr = np.bitwise_xor(hashed_arr, arr[:, j])
    return hashed_arr
```

# Visual Predictions and further Evaluation

```python
pred_choice = (preds > 0.5).astype("int")
tp = np.sum(np.logical_and(targets == 1, pred_choice == 1))
fn = np.sum(np.logical_and(targets == 1, pred_choice == 0))
fp = np.sum(np.logical_and(targets == 0, pred_choice == 1))
tn = np.sum(np.logical_and(targets == 0, pred_choice == 0))

precision = tp / (tp + fp)
recall = tp / (tp + fn)
f1 = 2*precision*recall / (precision + recall)
print(f1, precision, recall)
```

```python
#@title definition of visual analysis function
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
i=0
def explore2(points, prediction, label, mode="pointcloud", binary_colormode="True"):

    prediction = prediction.copy()

    # explore results with pointcloud
    if mode == "pointcloud":
        if binary_colormode:
            prediction[np.logical_and(label == 1, prediction >= 0.5)] = 4  # true positive
            prediction[np.logical_and(label == 1, prediction <= 0.5)] = 2  # false negative
            prediction[np.logical_and(label == 0, prediction >= 0.5)] = 3  # false positive
            prediction[np.logical_and(label == 0, prediction <= 0.5)] = 1  # true negative
        
        df = pd.DataFrame(data=np.column_stack((points, label, prediction)), columns=["x", "y", "z", "label", "prediction"])
        # create vector of sizes (two sizes for tree and non-tree points)
        size = df.iloc[:, -2] * 0.02 + 1

        fig = px.scatter_3d(df, x='x', y='y', z='z',
                            color='prediction'
                            , size=size, opacity=0, size_max=5, template="plotly_dark")
        fig.update_layout(paper_bgcolor='rgba(0,0,0,0)',
                           plot_bgcolor='rgba(0,0,0,0)'
                    )

        # tight layout
        fig.update_layout(margin=dict(l=0, r=0, b=0, t=0))

        fig.show()
        fig.write_html("/content/file" + str(i) + ".html")

    # explore results with histogram
    if mode == "histogram":
        tp = sum(np.logical_and(label == 1, prediction >= 0.5))
        fn = sum(np.logical_and(label == 1, prediction <= 0.5))
        fp = sum(np.logical_and(label == 0, prediction >= 0.5))
        tn = sum(np.logical_and(label == 0, prediction <= 0.5))
        print(f"{tp / len(label)*100:.2f} are true positive")  # true positive
        print(f"{fn / len(label)*100:.2f} are false negative")  # false negative
        print(f"{fp / len(label)*100:.2f} are false positive")  # false positive
        print(f"{tn / len(label)*100:.2f} are true negative")  # true negative
        
        fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True)
        axs[0].hist(prediction)
        axs[1].hist(label)
        fig.show()
```

```python
# plot binary predictions (yellow = true positive, red = false positive, blue = true negative, purple = false negative)
i=i+1
explore2(allpoints[::10], preds.astype("float")[::10], targets.astype("float")[::10], mode="pointcloud", binary_colormode=False)
```
